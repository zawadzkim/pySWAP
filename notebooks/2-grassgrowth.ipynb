# Grass growth example

In this example we will load the testcase model from pyswap test case module and we will look at how we can display the data. 

```python
import pyswap as ps
from pyswap import testcase

grassmodel = testcase.get('grassgrowth')
result = grassmodel.run('./', silence_warnings=True)
```

## Interpreting the results

```python
from pyswap import testcase
import seaborn as sns
obs_yield = testcase.load_observations('grassgrowth', parse_dates=True, index_col='date')

sns.lineplot(result.output[['PMOWDM', 'MOWDM']])
sns.scatterplot(obs_yield)
```
